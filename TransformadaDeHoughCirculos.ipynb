Deteção de circulos

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

def detect_circles(img, img_color):
    img_blur = cv2.GaussianBlur(img, (3, 3), 0)

    circles = cv2.HoughCircles(img_blur, cv2.HOUGH_GRADIENT, dp=1, minDist=260,
                               param1=60, param2=30, minRadius=150, maxRadius=200)

    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i in circles[0, :]:
            cv2.circle(img_color, (i[0], i[1]), i[2], (0, 255, 0), 2)
            cv2.circle(img_color, (i[0], i[1]), 2, (0, 0, 255), 3)

    return img_color

def main():
    img_color = cv2.imread('/content/economia-reciclagem-reciclaveis-mercado-mineiro-1713407444.jpeg', cv2.IMREAD_COLOR)

    if img_color is None:
        print("Erro ao carregar a imagem.")
        return

    img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

    img_with_circles = detect_circles(img_gray, img_color.copy())

    img_blur = cv2.GaussianBlur(img_gray, (3, 3), 0)
    img_canny = cv2.Canny(image=img_blur, threshold1=100, threshold2=200, apertureSize=3)

    cv2_imshow(img_gray)         # Imagem em escala de cinza
    cv2_imshow(img_canny)        # Imagem com bordas detectadas
    cv2_imshow(img_with_circles) # Imagem colorida com círculos

if __name__ == "__main__":
    main()
```
