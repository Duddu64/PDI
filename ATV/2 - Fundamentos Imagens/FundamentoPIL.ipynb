```python
from re import X
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def Translacao(img):
  matrix = (1, 0, 100, 0, 1, 50)
  t_img = img.transform(img.size, Image.AFFINE, matrix)
  matrix1 = (1, 0, 35, 0, 1, 45)
  t_img1 = img.transform(img.size, Image.AFFINE, matrix1)
  Plota(img, t_img)
  Plota(img, t_img1)

def Escala(img):
  l=int(img.width*1.5)
  m=int(img.width*2.5)
  o1 = img.resize((l, l))
  o2 = img.resize((m, m))
  Plota(img, o1)
  Plota(img, o2)


def Rotacao(img):
  o1 = img.rotate(45)
  o2 = img.rotate(90)
  o3 = img.rotate(100)
  Plota(img, o1)
  Plota(img, o2)
  Plota(img, o3)


def Plota(img, imgNew):
  fig, ax = plt.subplots(nrows=1, ncols=2, )
  ax[0].imshow(img, cmap='gray')
  ax[0].set_title("Imagem original")
  ax[1].imshow(imgNew, cmap='gray')
  ax[1].set_title("Nova imagem")
  plt.show()


def main():
  im1 = Image.open('/content/cameraman.tif')
  im2 = Image.open('/content/house.tif')
  im3 = Image.open('/content/lena_gray_512.tif')
  Translacao(im3)
  Rotacao(im3)
  Escala(im3)
  Translacao(im2)
  Rotacao(im2)
  Escala(im2)
  Translacao(im1)
  Rotacao(im1)
  Escala(im1)


if __name__ == "__main__":
  main()
```
