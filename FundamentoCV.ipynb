```python
from re import X
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

def Translacao(img):
  img_np = np.array(img)
  rows, cols = img_np.shape[:2]
  M = np.float32([[1, 0, 100], [0, 1, 50]])
  new_img = cv2.warpAffine(img_np, M, (cols, rows))
  M1 = np.float32([[1, 0, 35], [0, 1, 45]])
  new_img1 = cv2.warpAffine(img_np, M1, (cols, rows))
  Plota(img_np, new_img)
  Plota(img_np, new_img1)

def Escala(img):
  x1 = 1.5
  y1 = 1.5
  x = 2.5
  y = 2.5
  img_np = np.array(img)
  img_aumentada = cv2.resize(img_np, None, fx=x1, fy=y1, interpolation= cv2.INTER_LINEAR)
  img_aumentada1 = cv2.resize(img_np, None, fx=x, fy=y, interpolation= cv2.INTER_LINEAR)
  Plota(img, img_aumentada)
  Plota(img, img_aumentada1)

def Rotacao(img):
  img_np = np.array(img)
  rows, cols = img_np.shape[:2]
  M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
  rotate_45 = cv2.warpAffine(img_np, M, (cols, rows))
  M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 90, 1)
  rotate_90 = cv2.warpAffine(img_np, M, (cols, rows))
  M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 100, 1)
  rotate_100 = cv2.warpAffine(img_np, M, (cols, rows))
  Plota(img_np, rotate_45)
  Plota(img_np, rotate_90)
  Plota(img_np, rotate_100)


def Plota(img, imgNew):
  fig, ax = plt.subplots(nrows=1, ncols=2, )
  ax[0].imshow(img, cmap='gray')
  ax[0].set_title("Imagem original")
  ax[1].imshow(imgNew, cmap='gray')
  ax[1].set_title("Nova imagem")
  plt.show()


def main():
  im1 = Image.open('/content/cameraman.tif')
  im2 = Image.open('/content/house.tif')
  im3 = Image.open('/content/lena_gray_512.tif')
  Translacao(im3)
  Rotacao(im3)
  Escala(im3)
  Translacao(im2)
  Rotacao(im2)
  Escala(im2)
  Translacao(im1)
  Rotacao(im1)
  Escala(im1)

if __name__ == "__main__":
  main()
```
